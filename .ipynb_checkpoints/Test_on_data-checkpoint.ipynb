```python
import csv
import numpy as np
import matplotlib.pyplot  as plt

import torch
import torch.nn as nn

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_absolute_error

from verstack.stratified_continuous_split import scsplit # pip install verstack
```

```python
train_data = pd.read_csv('data/train.csv')
print(train_data)

X_train, X_test, y_train, y_test = scsplit(train_data, train_data['retweet_count'], stratify=train_data['retweet_count'], train_size=0.7, test_size=0.3)
eval_data = pd.read_csv('data/evaluation.csv')
```

```python
for i in train_data.keys()[:7]:
    print(i)
    print("min:", min(train_data[i]),"max: ", max(train_data[i]))
    plt.title("Nb of retweet as a function of " + i)
    plt.scatter(train_data[i],train_data["retweet_count"])
    plt.show()
```

```python
for i in train_data.keys()[:7]:
    print(i)
    print("min:", min(train_data[i]),"max: ", max(train_data[i]))
    plt.title("Nb of retweet as a function of " + i)
#     plt.scatter(train_data[i],train_data["retweet_count"])
    plt.scatter(train_data[i],np.log(train_data["retweet_count"]))
    plt.show()
```

```python
plt.plot([np.count_nonzero(np.array(train_data["retweet_count"]) == i) for i in range(1000)])
plt.show()

tot = 0
for i in range(20):
    C = np.count_nonzero(np.array(train_data["retweet_count"]) == i)
    tot += C
    print(i, " = ",C,", cumulative: ", tot/len(train_data))
```
